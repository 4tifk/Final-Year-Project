```python
# Library Import(numpy and matplotlib)
import numpy as np 
import matplotlib.pyplot as plot 
 
# Make a data definition
c1 = [1.0, 0.69, 1.0, 0.83]
c2 = [1.0, 0.84, 1.0, 0.79]
c3 = [1.0, 0.71, 1.0, 0.81]

bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:purple', 'tab:orange']
 
models = ['DT', 'GNB', 'MLP', 'SGD']
width = 1.0
 
# Create the first subplot
ax1 = plot.subplot(1, 3, 1)
ax1.bar(models, c1, width=width, color=bar_colors, align="center")
ax1.set_title('Recall')
ax1.set_ylabel('Value')
 
# Create the second subplot
ax2 = plot.subplot(1, 3, 2)
ax2.bar(models, c2, width=width, color=bar_colors, align="center")
ax2.set_title('Precision')

# Create the third subplot
ax3 = plot.subplot(1, 3, 3)
ax3.bar(models, c3, width=width, color=bar_colors, align="center")
ax3.set_title('F1 Score')
# Display the plot
plot.tight_layout()
plot.show()
```
